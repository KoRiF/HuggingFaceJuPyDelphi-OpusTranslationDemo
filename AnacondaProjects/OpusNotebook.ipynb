```python
####$ HuggingFace-based Opus translators $####


from transformers import pipeline

import logging
logging.disable(logging.WARNING)

import warnings
warnings.filterwarnings("ignore")

     
from transformers import AutoModelWithLMHead, AutoTokenizer


class OpusTranslator():
    def __init__(self, sharedState):
        self.state = sharedState
        self.translated = ""
        self.from_lang = ""
        self.to_lang = "" 
        self.split_by = "."        
        return
    
    
    def getState(self):
        return 
    
    def setup_cache(self):
            self.state.setTransformersOfflineFlag("1")
    def reset_cache(self):
            self.state.setTransformersOfflineFlag("0")
    def update_cache_policy(self):
            if self.state.updater:
                print(self.state.updater)
                self.state.updater(self.state.getTransformersOfflineFlag())
            
        
    def translate(self, source_text, lang_from = "en", lang_to = "ru"):       
        if (lang_from != self.from_lang) or (lang_to != self.to_lang):
            #######################
            #self.state.setTransformersOfflineFlag("0")
            self.reset_cache() 
            ########################
            model_tag = f"Helsinki-NLP/opus-mt-{lang_from}-{lang_to}"
            model = AutoModelWithLMHead.from_pretrained(model_tag)
            tokenizer = AutoTokenizer.from_pretrained(model_tag)
            pipeline_tag = f"translation_{lang_from}_to_{lang_to}"
            self.translator = pipeline(pipeline_tag, model=model, tokenizer=tokenizer)
        #################
        #os.environ["TRANSFORMERS_OFFLINE"]=self.state.getTransformersOfflineFlag()
        self.update_cache_policy()
        #################
        self.translated = ""
        sentences = source_text.split(self.split_by)
        for sent in sentences:
           translated = self.translator(sent)
           translation_text = translated[0]["translation_text"] 
           print(translation_text)
           self.translated += translation_text 
        ##################    
        #self.state.setTransformersOfflineFlag()  
        self.setup_cache()
        ##################
        return self.translated    
    


DEFAULT_FILENAME = r"c:\Users\KoRiF\Documents\Delphi Component Creation Notes.txt"

class SharedState():        
        def getFilename(self):
            return DEFAULT_FILENAME
        def getMode(self):
            return "file"
        def getSource(self):
            return ""
        def getLanguages(self):
            return "en", "uk"
        def getTransformersOfflineFlag(self):
            return "1" if self.cached else "0"
        def setTransformersOfflineFlag(self, flag = "1"):
            self.cached = (flag == "1")
        def get_flagpole(self):
            return self._flagpole
        def set_flagpole(self, value):
            self._flagpole = value
        def __init__(self): 
            self._flagpole = None
            self.flagpole = property(self.get_flagpole, self.set_flagpole)
            self.updater = None
             
        @classmethod
        def Instantiate(cls, kind):
            if kind == "Delphi":
                return DelphiSharedState()
            return SharedState()
        
class DelphiSharedState(SharedState):
        """{
        '''$
        }"""            
        def __init__(self):
            super().__init__()
            self.cached = True
        """{
        $'''
        }"""

        """%    
        from delphi_module import delphi_form
        def getFilename(self):
            return delphi_form.TranslFilename
        def getMode(self):
            return "file" if(delphi_form.TranslFilename) else ""
        def getSource(self):
            return delphi_form.TranslSource
        def getLanguages(self):
            return delphi_form.FromLang, delphi_form.ToLang
                
        def getTransformersOfflineFlag(self):
            self.cached = delphi_form.UseCachedTransformers
            return super().getTransformersOfflineFlag()
        def setTransformersOfflineFlag(self, flag = "1"):
            super().setTransformersOfflineFlag()       
            delphi_form.UseCachedTransformers = self.cached     
        %"""

              
```
